# Random Forest

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df1=pd.read_csv(r"C:\Users\DELL\Downloads\C1_Ionosphere.csv")
df1
```

```python
df1.describe()
```

```python
df1.info()
```

```python
g = {"g":{"g":1,"b":2}}
df1 = df1.replace(g)
df1
```

```python
df1["g"].value_counts()
```

```python
x = df1.drop("g",axis=1)
y = df1["g"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.40)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
```

```python
rf = RandomForestClassifier()
```

```python
params = {"max_depth":[1,2,3,4,5],
         "min_samples_leaf":[2,4,6,8,10],
         "n_estimators":[1,3,5,7]
        }
```

```python
from sklearn.model_selection import GridSearchCV
gs = GridSearchCV(estimator=rf,param_grid=params,cv=2,scoring='accuracy')
gs.fit(x_train,y_train)
```

```python
rf_best = gs.best_estimator_
rf_best
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(40,40))
plot_tree(rf_best.estimators_[4],feature_names=None,class_names=['Yes','No'])
```

```python
x = df2.drop("Marital Status",axis=1)
y = df2["Marital Status"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.40)
```
